```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
tips = sns.load_dataset('tips')
```

```python
# head what species are there?
# pairplot and tell me if you see correlation
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species', palette='prism') #prism, coolwarm, rainbow
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# draw the same for iris
sns.lmplot(x='petal_width', y='petal_length', data=iris)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
# survived: who survived the sanking of the titanic
# pclass: ticket class
# sex: passenger's gender
# age: passenger's age
# sibsp: # of spouses/sibings
# parch: # of parents/children
# fare: passenger fare
# embarked: port of embarkation
# class: class for royalty or commoner
# who: gender
# adult_male: whether they are an adult male or not
# deck: which deck they came from
# embark_town: which town/city they come from
# alive: check whether they are alive or not after the sinking
# alone: whether they board the titanic alone or with someone else
```

```python
sns.boxplot(x='class', y='age', data=titanic, palette='prism')
# the correlation is that the older the passenger, the higher the class except for elders
```

```python
sns.heatmap(titanic.corr())
```
