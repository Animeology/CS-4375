```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 39

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
df['age']= df[['age', 'pclass']].apply(impute_age, axis=1) 
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop(['cabin', 'boat',  'body', 'home.dest'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.describe().transpose()
```

```python
sex = pd.get_dummies(df['sex'], drop_first=True)
embarked = pd.get_dummies(df['embarked'], drop_first=True)
```

```python
df.drop(['sex', 'embarked', 'name', 'ticket'], axis=1, inplace=True)
```

```python
df.info()
```

```python
df = pd.concat([df, sex, embarked], axis=1)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('survived', axis=1), df['survived'], test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression(max_iter=1000)
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
# Exam review
# Lost functions and we need to minimize these
# Mean Absolute Error (MAE): is the mean of the absolute value of the errors
    # is the easiest to understand, because it is the average error
# Mean Squared Error (MSE): is the mean of the squared errors
    # is more popular than MAE, because MSE "punishes" large errors, which tends to be useful in the real world
# Root Mean Squared Error (RMSE): is the square root of the mean of the squared errors 
    # is even more popular than (MSE), because RMSE is interpretable in y units
from sklearn import metrics
print('MAE', metrics.mean_absolute_error(y_test, predictions))
print('MSE', metrics.mean_squared_error(y_test, predictions))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, predictions)))

# Run this for some small dataset to understand the concept behind these
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(df.drop('target', axis=1), df['target'],test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
X_train
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.fit_transform(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
plot_confusion_matrix(log_model, scaled_X_test, y_test)
```
